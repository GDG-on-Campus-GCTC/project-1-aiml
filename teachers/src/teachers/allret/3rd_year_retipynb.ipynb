```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

loader = PyMuPDFLoader("/content/SML.pdf")
pages = loader.load()

total_pages = len(pages)

for page_num, page in enumerate(pages):
    page.page_content = page.page_content.strip()
    page.metadata.update({
        "YEAR": 1,
        "subject": "SML or Statistics for machine learning",
        "page": page_num + 1,
        "total_pages": total_pages
    })

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunks = splitter.split_documents(pages)


for i, chunk in enumerate(chunks):
    chunk.metadata["chunk_id"] = i

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# 7. Create FAISS vector store
vectorstores = FAISS.from_documents(chunks, embeddings)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS


def yers(pdf_path: str, name: str):

    loader = PyMuPDFLoader(pdf_path)
    pages = loader.load()
    total_pages = len(pages)

    for page_num, page in enumerate(pages):
        page.page_content = page.page_content.strip()
        page.metadata.update({
            "YEAR": 2,
            "subject": name,
            "page": page_num + 1,
            "total_pages": total_pages
        })

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )
    chunks = splitter.split_documents(pages)

    for i, chunk in enumerate(chunks):
        chunk.metadata["chunk_id"] = i

    embedding = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstores.add_documents(chunks)

    return vectorstores
```
