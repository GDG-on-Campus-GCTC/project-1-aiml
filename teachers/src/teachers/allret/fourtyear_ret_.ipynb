```python
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS


def yers(pdf_path: str, name: str):

    loader = PyMuPDFLoader(pdf_path)
    pages = loader.load()
    total_pages = len(pages)

    for page_num, page in enumerate(pages):
        page.page_content = page.page_content.strip()
        page.metadata.update({
            "YEAR": 2,
            "subject": name,
            "page": page_num + 1,
            "total_pages": total_pages
        })

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=100
    )
    chunks = splitter.split_documents(pages)

    for i, chunk in enumerate(chunks):
        chunk.metadata["chunk_id"] = i

    embedding = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vectorstore.add_documents(chunks)

    return vectorstore
```

```python
yers("/content/Mobile Application Development.pdf","MAD OR Mobile Application Development")
yers("/content/Software Testing Methodologies.pdf","STM or Software Testing Methodologies")
yers("/content/bda.pdf","bda or Big Data Analysis")
yers("/content/ml machine learning.pdf","machine learnig or ml ")
```
